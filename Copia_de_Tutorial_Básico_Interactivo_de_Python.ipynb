# 1. Asignación de variables

Los principales tipos de datos son números enteros (int), números decimales (float), cadenas de texto (str) y valores booleanos (verdadero o falso, te cuento más sobre esto más adelante).

Para realizar asignaciones utiliza nombres de variables identificativos, el símbolo = y el valor que quieras asociar a la variable.

```python
x = 10     # enteros
y = 0.5    # decimales
z = 'hola' # texto
b = True   # booleanos
print(x)
print(y)
print(z)
print(b)
```

Las principales colecciones de datos son las listas (modificables), las tuplas (no modificables), los conjuntos (que no admiten repetidos) y los diccionarios (que guardan pares clave-valor).

```python
tupla = (1, 2, 3)
lista = [4, 5.5, 'a6']
conjunto = ('a', 'b', 'c')

print(tupla)
print(lista)
print(conjunto)

diccionario = {10: 1.1, 20: 1.0, 30: 1.3}
diccionario[40] = 1.5  # asignando un valor (1.5) a una nueva clave (40)
print(diccionario)

x, y, z = 1, 2, 3  # asignación múltiple
print('Tras la asignación múltiple: ')
print(x)
print(y)
print(z)

print('Tras intercambiar valores: ')
x, y = y, x  # intercambio de valores
print(x)
print(y)

print('Tras desempaquetar una tupla en variables: ')
x, y, z = (10, 20, 30)
print(x)
print(y)
print(z)
```

# 2. Acceso a valor de variables

Cuando ya hemos hecho asignaciones y tenemos variables con valores, podemos acceder a esos valores utilizando el nombre de las variables en otras expresiones.

Además, podemos acceder a los elementos de una lista o una tupla a través de su índice (número de posición, comenzando en 0) indicándolo entre corchetes \[ y \].

Para acceder a los valores de un diccionario utilizaremos la clave asociada a cada valor, de nuevo entre corchetes.

También podemos acceder a los atributos de una clase utilizando el nombre de la clase, un punto y a continuación el nombre del atributo. De igual manera, podemos acceder a los atributos de un objeto o instancia utilizando el nombre del objeto, un punto y a continuación el nombre del atributo.


```python
lista = [10, 20, 30, 40, 50]
valor = lista[2]
print(valor)

diccionario = {10: 5, 20: 7, 30: 9}
valor = diccionario[20]
print(valor)
valor = diccionario.get(40, '0')  # valor por defecto 0 si clave (40) no existe

class Clase:

    atributo_de_clase = 10

    def __init__(self):
        self.atributo_de_instancia = 20


print(Clase.atributo_de_clase)

instancia = Clase()
print(instancia.atributo_de_instancia)
```

# 5. Gestión de excepciones

Para manejar errores y excepciones tienes que usar la construcción try - except - finally. De no hacerlo, el código interrumpe su ejecución en caso de surgir algún error.

Algunos de los errores más comunes son los siguientes:
* SyntaxError: sintaxis incorrecta
* ValueError: valor argumento no válido
* ZeroDivisionError: división por cero
* NameError: variable no definida
* ModuleNotFoundError: modulo no existe
* FileNotFoundError: fichero no existe
* IndexError: índice no existe
* KeyError: clave no existe





```python
# El siguiente código genera un error

a = 10
b = 0

print(a/b)  # no se puede dividir por cero
```

```python
# Vamos a controlar el error para que no se interrumpa la ejecución

a = 10
b = 0

try:
    print(a/b)
except ZeroDivisionError:
    print('Lo siento, no puedo dividir por cero.')
finally:
    print('Bloque try-except finalizado')  # esto, que es opcional, se ejecuta siempre, haya error o no
    
```

# 7. Listas

Podemos hacer muchas cosas con las listas, pues son secuencias muy potentes. Veamos algunas:

```python
lista = [5, 6, 7]

lista.append(8)  # añadimos un elemento al final
print(lista)

lista.insert(1, 10)  # insertamos un elemento en la posición que queramos
print(lista)

lista.remove(6)  # eliminamos un elemento
del lista[0]     # eliminamos una posición
print(lista)

lista.clear()  # vacíamos la lista
print(lista)

lista = [5, 6, 7]
elementos = [8, 9]

lista.extend(elementos)  # añadimos al final los elementos de otra lista
print(lista)

lista.reverse()  # invertimos la lista (le damos la vuelta)
print(lista)

print(len(lista))  # obtenemos el número de elementos (o longitud) de una lista

print(lista.index(7))  # obtenemos el índice de un elemento

```

# 8. Condicionales

Las condicionales nos permiten ejecutar un determinado fragmento de código si se cumple una determinada condición (rama if).

Además, podemos indicar que, si no se cumple dicha condición, se ejecute otro código (rama else).

Finalmente, se permite encadenar varias ramas ifs con distintas condiciones (elif) y una rama else si no se cumple ninguna de las condiciones indicadas. Las ramas aquí son excluyentes.

```python
a = 10
b = 20

if a < 15:
    print('a es menor que 15') # esto se ejecutará ya que se cumple la condición


if a > 15:
    print('a es mayor que 15') # esto no se ejecutará ya que no se cumple la condición

if a > 30:
    print('a es mayor que 30') # esto no se ejecutará ya que no se cumple la condición
else:
    print('a no es mayor que 30') # esto sí se ejecutará ya que no se cumple la condición


if b > 10:
    print('b es mayor que 10')  # esto se ejecutará ya que se cumple la condición
else:
    print('b no es mayor que 10')  # esto no se ejecutará ya que se cumple la condición


# a continuación se comprobaran las condiciones secuencialmente y se ejecutara 
# solo el fragmento de código de la condición que se cumpla (condiciones excluyentes)

if a < 10:
    print('a es menor que 10')
elif a > 12:
    print('a es mayor que 12')
elif b > 15:
    print('b es mayor que 15')  # esta es la rama que se ejecuta
elif b == 20:
    print('b es igual a 20')  # aunque la condición de esta rama es cierta no se ejcuta
                              # pues ya se ejecutó otra rama antes
else:
    print('no se ha cumplido ninguna de las condiciones anteriores')
```

# 9. Booleanos y comparaciones

Ahora que ya sabes cómo hacer condicionales, vamos a revisar los valores booleanos y las principales posibles comparaciones y comprobaciones que se pueden hacer en Python que devuelven valores booleanos.

Los valores booleanos, son True y False (escritos con inicial mayúscula) y cuyo significado es "cierto" y "falso" respectivamente.

Comparaciones y comprobaciones principales:
* Mayor: >
* Mayor o igual: >=
* Menor: <
* Menor o igual: <=
* Igual: ==
* Diferente: !=
* Pertenencia: in

Puedes negar un valor booleano o condición para obtener el valor contrario con la palabra reservada "not".

```python
a = 2
b = 3

print('Comparaciones simples:')
print(a > b)
print(a >= b)
print(a < b)
print(a <= b)
print(a == b)
print(a != b)

print('Comprobación de pertenencia:')
lista = [1, 3, 5]
print(a in lista)
print(b in lista)

print('Negación con not')
print(not True)
print(not False)
print(not a > b)
print(not a == b)
```

Más información sobre booleanos, comparaciones, condicionales, etc.: https://www.codigopiton.com/condicionales-en-python-mucho-mas-que-if-else/

Puedes combinar distintas comparaciones para crear una más compleja con las palabras reservadas "and" y "or".

Cuando usas and entre dos comparaciones o condiciones, ambas deben ser True para que la condición combinada sea True.

Cuando usas or entre dos comparaciones o condiciones, al menos una de ellas tiene que ser True para que la condición combinada sea True.

```python
a = 2
b = 3
print(b >= 3 and a == 2)
print(a < 10 and b < 2)
print(a == 2 or b == 0)
print(a < 0 or b < 0)
```

# 10. Comprensión de listas

En Python podemos generar listas "al vuelo" utilizando un bucle for que devuelva un elemento e en cada iteración, aplicando un cálculo o expresión sobre e y encerrándolo todo entre [corchetes](https://www.codigopiton.com/parentesis-corchetes-llaves-en-python/). 

De esta manera se genera una lista con todos los elementos generados a partir de e.

Ejemplos:

```python
lista = [10, 20, 30]

lista_negada = [ -e for e in lista ]
lista_doble = [ e*2 for e in lista]

print(lista_negada)
print(lista_doble)
```

# 12. Funciones

Tanto para estructurar nuestro código así como para favorecer la reutilización del mismo se utilizan funciones.

Las funciones nos permiten encapsular un fragmento de código y darle un nombre.

Para definir una función usamos la palabra reservada "def" a continuación el nombre de la función y los parámetros que necesitemos entre paréntesis. La función puede devolver uno o varios valores si lo necesita (o ninguno).

Después se puede invocar una función creada a través de su nombre.

```python
def suma(a, b):
    return a + b


# min y max son funciones integradas en Python que nos devuelven
# los valores mínimo y máximo de un iterable (como las listas o las tuplas)
def min_y_max(lista):
    return min(lista), max(lista)


# ahora vamos a usar estas funciones
valor = suma(10, 15)
print(valor)

minimo, maximo = min_y_max([30, 10, 50, 40, 20])
print(f'min y max son {minimo} y {maximo}')
```

# 13. Bucles

Los bucles son un mecanismo de control de flujo que nos permite repetir un fragmento de código un número determinado de veces.

Hay dos tipos:


1.   while, que nos permite controlar mediante una condición el número de repeticiones.
2.  for, que nos permite iterar por los elementos de una colección de manera sencilla y cómoda.

El código que queremos repetir, también llamado "cuerpo del bucle" debe indentarse tras el for o el while.



```python
# ejemplo de while
a = 0

while (a < 10):  # mientras a sea < 10 el cuerpo del bucle se repetirá
    print(a)
    a = a + 1  # sumamos 1 a la variable a para asegurarnos de que la condición del bucle será False en algún momento

# ejemplo de for
lista = [10, 20, 30, 40]

for elemento in lista:
    print(elemento)

```

Dentro de los bucles puedes utilizar las sentencias de control break y continue:
* break termina anticipadamente el bucle
* continue interrumpe la vuelta actual del bucle y continúa con la siguiente.

# 14. Notación de porciones

Cuando tenemos una colección de elementos indexados por posición (como una lista o una tupla), podemos acceder a una porción concreta utilizando la notación de porciones.

Se utilizan los corchetes entre los que indicamos tres números separados por el carácter ':'. De esta manera

lista\[ini:fin:paso\]

Donde:
* ini es la posición del elemento primero en nuestra porción.
* fin es la posición del elemeto último (no incluido) de nuestra porción (el último incluido será el correspondiente a la posición fin -1).
* paso (optativo) indica la manera en la que tomaremos la porción: si es un número positivo tomaremos los elementos en el orden en el que aparecen, si es negativo los tomaremos en orden inverso (fin tiene que ser anterior a ini). Si la magnitud es 1, tomaremos todos los elementos, si es 2, tomaremos uno y saltaremos otro, si es 3, tomaremos uno y saltaremos 2, etc.

Si dejamos ini o fin sin indicar se interpreta que la porción se toma desde el comienzo o el final respectivamente.

Veamos algunos ejemplos:


```python
lista = [10, 20, 30, 40, 50, 60]

print(lista[2:5])  # tomamos del 2 al 5 (sin incluir el 5)
print(lista[4:])  # tomamos del 4 hasta el final
print(lista[:2])  # tomamos desde el principio hasta el 2 (sin incluir)
print(lista[1:6:2]) # tomamos 1 de cada 2 elementos desde el 1 hasta el 6 (sin incluir)
print(lista[4:1:-1])  # tomamos en orden inverso desde el 4 al 1 (sin incluir)
print(lista[:])  # tomamos desde el principio hasta el final (es como hacer una copia de la lista)
print(lista[::-1])  # tomamos desde el final hasta el principip (orden inverso) 
```

# Ejercicio final: juego de adivinar un número entre 1 y 10

Para finalizar este tutorial, voy a realizar un pequeño juego en el que el usuario tiene que adivinar un número generado al azar entre 1 y 10 y en el que solo tendrá 3 intentos para hacerlo.

Para esto voy a utilizar muchas de las cosas contadas arriba. Además, comentaré el código para que todo quede perfectamente claro. Léelo con calma intentando comprender cada línea.

Y, por supuesto, ¡puedes ejecutarlo para jugar tú!


```python
import random  # necesitamos esta librería para poder generar un número al azar

numero_secreto = random.randint(1, 10)  # esto nos genera un número aleatorio entre 1 y 10
oportunidades = 3  # el número de oportunidades que nos quedan
acertado = False  # esta variable nos servirá para saber si el usuario ha acertado

print('* Bienvenido a nuestro pequeño juego del Tutorial Básico Interactivo de Código Pitón')
print()  # imprimimos por pantalla una línea en blanco
print('Acabo de pensar un número secreto entre 1 y 10. Tienes tres oportunidades para')
print('adivinarlo. Te iré guiando un poco hacia la solución. ¡Suerte!')
print()

# Hacemos un bucle mientras queden oportunidades y mientras no se haya acertado el número secreto
while oportunidades > 0 and not acertado:
    
    # Pedimos un número al usuario (y transformamos lo que escriba a entero mediante la función int)
    # ¡Cuidado! Si lo que escribe el usuario no es un entero el programa generará un error
    # El error puede ser controlado con try-except como vimos más arriba
    numero = int(input('Dime un número entre 1 y 10: '))

    # Ahora vamos a hacer la comprobación
    # e indicamos al usuario si el número secreto es mayor, menor o si lo ha averiguado
    if numero < numero_secreto:
        print('Te has quedado corto, el número secreto es más grande.')
    elif numero > numero_secreto:
        print('Te has pasado, el número secreto es más pequeño.')
    else:
        print(f'¡Enhorabuena! Has acertado en el intento {4-oportunidades}.')
        acertado = True  # ponemos acertado a cierto para que se detenga el bucle

    if not acertado:  # si no ha acertado indicamos las oprtunidades que le quedan
        oportunidades = oportunidades - 1  # restamos 1 a las oportunidades
        print(f'Te quedan {oportunidades} oportunidades.')

    print()

# cuando termine el bucle de juego mostramos el número de adinivinar
# y nos despedimos del usuario

if not acertado:  # si no se acierta el número se muestra por pantalla
    print(f'* El número secreto era {numero_secreto}.')

print('* ¡Gracias por jugar! ¡Hasta pronto!') 

```

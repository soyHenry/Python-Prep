```python
def factorizar_numero(numero):
    if (type(numero)!=int):
        return None
    if (numero < 1):
        return None
    lista_primos = []
    i = 2
    while(i < (numero / 2)):
        primo = True
        j = 2
        while (j < i):
            if (i % j == 0):
                primo = False
                break
            j+=1
        if (primo):
            lista_primos.append(i)
        i+=1
    lista_factores = []
    lista_exponentes = []
    for i in lista_primos:
        j = 0
        while (numero % i == 0):
            numero = numero / i
            if (j == 0):
                lista_factores.append(i)
            j+=1
        if (j>0):
            lista_exponentes.append(j)
    return lista_factores, lista_exponentes
```
